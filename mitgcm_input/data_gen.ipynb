```python
import numpy as np
import sys
accuracy='float64'

problem_size = 10**4

def writefield(name, field):
    doByteSwap = False
    if sys.byteorder=='little': doByteSwap = True

    (field.astype('float64').byteswap(doByteSwap)).tofile(name)

grid_len = int(np.sqrt(problem_size))
ones2d = np.ones((grid_len,grid_len))

ice = ones2d * 1
snow = ones2d * 0.1
area = ones2d * 0.7

h = ones2d * -1000
uwind = ones2d * 5
vwind = ones2d * 5
uo = ones2d * 0.1
vo = ones2d * 0.1

writefield('bathy_3c_8000.bin',h)
writefield('Uwindfield_shifted_8000.bin',uwind)
writefield('Vwindfield_shifted_8000.bin',vwind)
writefield('uVel_8000.bin',uo)
writefield('vVel_8000.bin',vo)
writefield('ice_file.bin', ice)
writefield('snow_file.bin', snow)
writefield('area_file.bin', area)
```
